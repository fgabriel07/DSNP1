## Obtenção dos Dados

Os dados que serão analisados neste dataset, foram obtidos do portal ISPDados, página de Dados Abertos do Instituto de Segurança Pública do Estado do Rio de Janeiro.

As estatísticas divulgadas no portal são construídas a partir dos Registros de Ocorrência (RO) lavrados nas delegacias da Secretaria de Estado de Polícia Civil do Rio de Janeiro (SEPOL), além de informações complementares de órgãos específicos da Secretaria de Estado de Polícia Militar do Rio de Janeiro (SEPM).

Antes de serem consolidados no ISP, os RO são submetidos ao controle de qualidade realizado pela Corregedoria Geral de Polícia (CGPOL) da Secretaria de Estado de Polícia Civil. As estatísticas produzidas baseiam-se na data em que foi confeccionado o Registro de Ocorrência.

No caso do dataset a ser utilizado aqui, o arquivo csv pode ser baixado diretamente [neste link](https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/violencia_rio.csv)

### Crimes Contra o Patrimônio - Veículos

Vamos dedicar essa seção do projeto para investigar dados em relação aos crimes de roubo e furto de veículos.

Algumas das variáveis que não possuem valores ausentes, são:

* `roubo_veiculo`;
* `furto_veiculos`;
* `recuperacao_veiculos`.

Sendo assim, podemos utilizar as informações presentes no dataset para entender a situação desse tipo de crime, buscando encontrar resultados pertinentes em relação à situação e ao combate desses crimes.

```
# média dos veículos roubados
veiculos_roubados = '{:.1f}'.format(df.roubo_veiculo.mean())

# média dos veículos furtados
veiculos_furtados = '{:.1f}'.format(df.furto_veiculos.mean())

# relação entre veículos roubados e veículos furtados
furtoxroubo = '{:.2%}'.format(df.furto_veiculos.mean() / df.roubo_veiculo.mean())

# média dos veículos recuperados
veiculos_recuperados = '{:.1f}'.format(df.recuperacao_veiculos.mean())

# porcentagem dos veículos que são recuperados
porcentagem = '{:.2%}'.format(df.recuperacao_veiculos.mean() / (df.roubo_veiculo.mean() + df.furto_veiculos.mean()))

print ('• Média de Veículos roubados por mês: {};\n'.format(veiculos_roubados),
      '\n• Média de Veículos furtados por mês: {};\n'.format(veiculos_furtados),
      '\n• Em média, há uma taxa de {}'.format(furtoxroubo), ('a mais de veículos roubados em relação aos que são furtados;\n'),
      '\n• Média dos Veículos recuperados por mês: {};\n'.format(veiculos_recuperados),
      '\n• {} dos Veículos roubados ou furtados, são recuperados.'.format(porcentagem))
```

Podemos observar a grande quantidade de ocorrências para crimes de roubo e furto de vículos, explanando a preocupante situação da violência. E, ainda mais alarmante se levarmos em consideração que, muitas vezes, esse tipo de crime está ligado a outros tipos de crime. 

Além de uma média pouco satisfatória em relação aos veículos que são recuperados, vale pontuar a relação entre a quantidade de veículos roubados e veículos furtados, pois, se tratando de roubo, a forma como o crime é executado, pode não apenas gerar traumas às vítimas, como ainda podem acabar se tornando crimes de latrocínio.
